```python
import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

data = {'Person': people,'Age': age, 'Height': height, 'Weight': weight}

# Create a DataFrame
health_care_dataset = pd.DataFrame(data)

# Display the dataset
print(health_care_dataset)
```
